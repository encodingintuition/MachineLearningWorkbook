```python
# create some data
X, y = make_blobs(n_samples = 20, centers = 2, random_state = 22)
```

```python
# plot data in scatter 
plt.scatter(X[:, 0], X[:, 1], c = y, s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)

plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Two features $X1, X2$ plotted colored by class');
```

```python
# create model and Fit
clf = LogisticRegression()
clf.fit(X,y)
```

```python
# circles toy data
X, y = make_circles(n_samples=50, noise = .2, factor = 0.2, random_state = 24)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y, s=50,
           cmap="RdBu", vmin=-.2, vmax=1.2,
           edgecolor="white", linewidth=1)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('A Non Linear Decision Boundary');
```

```python
# Polynomial instance
poly = PolynomialFeatures()
# Logistic Regression instance
clf = LogisticRegression()
```

```python
# fit polynomial to X
X_poly = poly.fit_transform(X)
```

```python
# train test split 
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 11)
```

```python
# fit logistic regression
clf.fit(X_train, y_train)
```

```python
"""
  Accuracy score
"""

clf.score(X_test, y_test)
```

```python
# cross validation
cross_val_score(clf, X_train, y_train)
```
